```python
import glob
import pandas as pd
from datetime import datetime
import os
```

```python
path_enviados = os.getcwd() +'\\' + 'enviados\\backup'
```

```python
path_enviados
```

```python
all_files = glob.glob(path_enviados + "/*"+".csv")

li = []

for filename in all_files:
    try:
        df = pd.read_csv(filename, index_col=None, header=0, delimiter ="\s+|;|,")
        #df.drop(columns=['FLAG_REDISTRIBUICAO','Unnamed'], inplace=True)
        df["data"]=datetime.fromtimestamp(os.path.getmtime(filename)).strftime('%Y%m%d')
        df.rename(columns={',ID_LEAD':'ID_LEAD'}, inplace=True)
        #df_final=pd.concat([df_inicio,df])
        li.append(df)
        
    except: pass
 
```

```python
        
try:     
    enviados = pd.concat(li, axis=0, ignore_index=True)
    #enviados.drop_duplicates(subset=['data', 'ID_LEAD'], inplace=True)
    #lista de corretores que j√° trabalharam hoje por id do lead
    enviados_2 = enviados
    enviados_2['ID_LEAD'] = enviados_2[',ID_LEAD'].apply(lambda x: x.split(',')[1]).astype(float)
except: pass
```

```python
enviados_2.drop(columns=['FLAG_REDISTRIBUICAO'])
```

```python
enviados_2.to_csv('historico_match.csv', index=False)
```
